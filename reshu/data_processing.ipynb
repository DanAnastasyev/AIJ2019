```python
import re
```

```python
def clean_html(raw_html):
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, ' ', raw_html.replace('\xad', '').replace('\xa0', ' '))
    return cleantext
```

```python
dataset = []
from collections import Counter
s = Counter()
for el in data:
    res = {}
    try:
        res['num'] = int(re.match('Задани[ея] Д?(\d+)[^\d].*', el['num']).groups()[0])
        res['answer'] = clean_html(el['answer']).strip(' .') if el['answer'] is not None else None
        res['body'] = [clean_html(part) for part in el['body']]
        res['text'] = clean_html(el['text']) if el['text'] is not None else el['text']
        res['solution'] = clean_html(el['solution'])
    except (AttributeError, TypeError):
        if ('Д' in el['num']):
            s[el['num']] += 1
        else:
            print(el)
            break
    else:
        dataset.append(res)
print(s)
```

```python
from copy import deepcopy
```

```python
processed = []
skip = set()
for row in dataset:
    num = row['num']
    t = deepcopy(template[num - 1])
    q_type = t['question']['type']
    answer = row['answer']
    variants = answer.lower().split('|') if answer else []
    body = row['body']
    if q_type == 'text':
        pass
    elif q_type in ['choice', 'multiple_choice']:
        choice_type = t['question']['choices'][0].keys()
        variants = [list(el) for el in variants]
        if 'text' in choice_type:
            body = []
            options = []
            for p in row['body']:
                p = p.strip()
                match = re.match('(\d+)[).].*', p)
                if match:
                    options.append({'id': match.groups()[0], 'text': p})
                else:
                    body.append(p)
            if not options:
                #print(row)
                continue
            t['question']['choices'] = options
        elif 'placeholder' in choice_type or 'link' in choice_type:
            text = '\n'.join(body) + ' ' + (row['text'] or '')
            choice_type = 'placeholder' if 'placeholder' in choice_type else 'link'
            matches = re.findall('\(\d+\)', text)
            options = [{'id': int(match[1:-1]), choice_type: match} for match in matches]
        else:
            print(row)
            break
    elif q_type == 'matching':
        body = []
        left = []
        right = []
        for p in row['body']:
            p = p.strip()
            if num == 26:
                matches = re.findall('\([^\d]\)', p)
                left += [{'id': match[1:-1], 'placeholder': match} for match in matches]
            match = re.match('([^\d])[).].*', p)
            if match:
                left.append({'id': match.groups()[0], 'text': p})   
            else:
                match = re.match('(\d+)[).].*', p)
                if match:
                    right.append({'id': match.groups()[0], 'text': p})
                else:
                    body.append(p)
        if not options:
            continue
        t['question']['left'] = left
        t['question']['choices'] = right
        variants = [{k['id']: v for k, v in zip(left, variants[0])}]
    else:
        skip.add(num)
        continue
    t['solution'] = dict(correct_variants=variants)
    t['text'] = '\n'.join(body) + ' ' + (row['text'] or '')
    t['hint'] = row['solution']
    processed.append(t)  
print(skip)
```
